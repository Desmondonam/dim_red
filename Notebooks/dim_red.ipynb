```python
import pandas as pd

# Assuming df is your DataFrame
numerical_df = df.select_dtypes(include=['number'])
numerical_df.head()
```

```python
# Compute correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar_kws={'shrink': .8})
plt.title('Correlation Matrix')
plt.show()
```

Remove highly correlated features

```python
# Threshold for removing correlated features
threshold = 0.9

# Find pairs of correlated features
correlated_features = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname = corr_matrix.columns[i]
            correlated_features.add(colname)

# Drop the correlated features
df_reduced = numerical_df.drop(columns=correlated_features)

# Display the remaining features
print("Features retained after removing highly correlated features:")
print(df_reduced.columns)
```

# PCA
Standadize the data

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the remaining features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_reduced)
```

apply PCA

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce dimensionality
pca = PCA(n_components=0.95)  # Retain 95% of the variance
df_pca = pca.fit_transform(df_scaled)

# Check how much variance is explained by each component
explained_variance = pca.explained_variance_ratio_
print("Explained variance by each principal component:")
print(explained_variance)

# Total variance retained
print(f"Total variance retained: {explained_variance.sum():.2f}")
```

```python
# visualize PCA
# Scree plot to visualize the explained variance by each component
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
# 2D Scatter plot of the first 2 Principle components
# Scatter plot of the first two principal components
plt.figure(figsize=(10, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], alpha=0.7, edgecolors='w', s=100)
plt.title('2D Scatter Plot of First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Save the PCA Transformed data
# Convert PCA results to a DataFrame
df_pca_df = pd.DataFrame(df_pca, columns=[f'PC{i+1}' for i in range(df_pca.shape[1])])

# Save the PCA-transformed dataset
pca_file_path = '../Data/pca_transformed_data.csv'
df_pca_df.to_csv(pca_file_path, index=False)

print(f"PCA-transformed dataset saved at {pca_file_path}")
```
